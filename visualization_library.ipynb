```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.
import numpy as np
import pandas as pd

np.random.seed(42)

# Generate synthetic data
n_samples = 100
x = np.random.normal(5, 1, n_samples)  # Normally distributed data for x
y = 2 * x + np.random.normal(0, 1, n_samples)  # y is linearly related to x with some noise


data = pd.DataFrame({
    'X': x,
    'Y': y
})

```

```python
data
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(x='X', y='Y', data=data, color='blue', alpha=0.7)
sns.regplot(x='X', y='Y', data=data, scatter=False, color='orange')

plt.title('Scatter Plot of X vs Y')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```

```python
#2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

# Set the random seed for reproducibility
np.random.seed(42)

# Generate random numbers
n_samples = 1000
data = {
    'Random Numbers': np.random.normal(loc=0, scale=1, size=n_samples)  # Mean = 0, Std Dev = 1
}

# Create a DataFrame
df = pd.DataFrame(data)
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

# Set the random seed for reproducibility
np.random.seed(42)

# Generate correlated data
n_samples = 100
x1 = np.random.normal(0, 1, n_samples)
x2 = x1 * 0.5 + np.random.normal(0, 0.5, n_samples)  # x2 is correlated with x1
x3 = np.random.normal(0, 1, n_samples)
x4 = x1 * -0.3 + x3 * 0.7 + np.random.normal(0, 0.5, n_samples)  # x4 is correlated with x1 and x3

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})
```

```python
data
```

```python
correlation_matrix = data.corr()
correlation_matrix
```

```python
plt.figure(figsize=(8, 6))

sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix Heatmap')

plt.show()
```

```python
#2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories.
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)

fig = px.violin(df, x='Grade', y='Score')
fig.show()

```

```python
#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

df = df.groupby(['Month', 'Day'])['Sales'].sum().reset_index()

fig = px.imshow(df.pivot(index='Month',columns='Day',values='Sales'))
fig.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.show()
```

#BOKEH ASSIGNMENT:

(Use Bokeh for the visualization of the given questions)

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.sampledata.iris import flowers
from bokeh.plotting import figure,output_notebook,show
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

output_notebook()
x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)',
           width=800, height=400)

p.line(x, y, legend_label="sin(x)", line_width=2, line_color='blue')
```

```python
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
from bokeh.models import ColumnDataSource

output_notebook()
np.random.seed(42)
n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100
sizes = np.random.randint(5, 50, size=n)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)

data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

source = ColumnDataSource(data)

p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y',  width=800, height=400)

p.circle('x', 'y', size='sizes', color='colors', source=source, legend_field='colors', fill_alpha=0.6)

show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruits', y_axis_label='Counts',
           width=800, height=400)

p.vbar(x='fruits', top='counts', source=source, width=0.5,
       fill_color='blue', line_color='black')

show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram of Random Data",
           x_axis_label='Value', y_axis_label='Frequency',
           width=800, height=400)

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color='blue', line_color='black')

show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource # Import ColumnDataSource
from bokeh.palettes import Viridis256


output_notebook()

# Prepare data for Bokeh
x_coords, y_coords = np.meshgrid(x, y)
x_coords = x_coords.flatten()
y_coords = y_coords.flatten()
data_flat = data_heatmap.flatten()

# Create a ColumnDataSource with the data
source = ColumnDataSource(data=dict(x=x_coords, y=y_coords, value=data_flat)) # Create a ColumnDataSource

p = figure(title="Heatmap",
           x_axis_label='X', y_axis_label='Y',
           width=600, height=600)

mapper = linear_cmap(field_name='value', palette=Viridis256, low=data_flat.min(), high=data_flat.max())

# Refer to the columns in the source for x, y, and fill_color
p.rect(x='x', y='y', width=0.1, height=0.1, source=source, line_color='white', fill_color=mapper)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
